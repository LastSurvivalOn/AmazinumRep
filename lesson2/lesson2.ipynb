```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2023)
a = np.random.randint(1,100,20)
plt.plot(range(0, 20), a, c='blue')


min_v=np.argmin(a)
indexx1=0
indexx2=0

if min_v==0:
    indexx1=0
    indexx2=min_v+1
    while indexx2 < len(a) and a[indexx2] < a[indexx2 - 1]:
        indexx2 += 1
elif min_v==len(a)-1: 
    indexx2=len(a)-1
    indexx1=min_v-1
    while indexx1 >= 0 and a[indexx1] < a[indexx1 + 1]:
        indexx1 -= 1
else:
    indexx1=min_v-1
    indexx2=min_v+1
    while indexx2 < len(a) and a[indexx2] < a[indexx2 + 1]:
        indexx2 += 1
    while indexx1 >= 0 and a[indexx1] < a[indexx1 - 1]:
        indexx1 -= 1


x = range(indexx1, indexx2+1)
plt.plot (x, a[indexx1:indexx2+1], c='red')

plt.show()
```
