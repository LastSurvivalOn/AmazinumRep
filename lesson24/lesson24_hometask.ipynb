```python
import pandas as pd 
fn= 'voted-kaggle-dataset.csv'
df = pd.read_csv(fn)
```

```python
data=df['Description'].dropna()
len(data)
```

```python
from gensim import models
import gensim
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

three_words_pattern = r"\b\w{3,}\b"

stop  = list(set(stopwords.words('english')).union({'data', 'dataset','context','content'}))

vectorizer = CountVectorizer(
    min_df=20, 
    stop_words=stop,
    token_pattern=three_words_pattern) 
vectorizer.fit(data)
```

```python
print ('len of features = {:,}\n'.format(len(vectorizer.get_feature_names_out())))
print (vectorizer.get_feature_names_out()[:40])
```

```python
data_vectorized= vectorizer.transform(data)
print (data_vectorized)
```

```python
corpus = gensim.matutils.Sparse2Corpus(data_vectorized, documents_columns=False)
```

```python
id_map = dict((v, k) for k, v in vectorizer.vocabulary_.items())
id_map
```

```python
ldamodel = gensim.models.ldamodel.LdaModel (corpus, num_topics=6, id2word=id_map, passes=25, random_state=2024)
```

```python
ldamodel.print_topics(num_topics=6,num_words=100)
```

```python
topics_names= ['ML', 'Sports', 'Eductaion', 'Social', 'Dataset', 'Internet']
```

```python
doc_vectorized= vectorizer.transform(new_doc)
new_doc_corpus = gensim.matutils.Sparse2Corpus(doc_vectorized, documents_columns=False)
doc_topics = ldamodel.get_document_topics(new_doc_corpus)
list(doc_topics)
```

```python
def elicit_topic_name(doc_topics):    
    topic_probabilities = [max(probabilities, key=lambda x: x[1]) for probabilities in doc_topics]
    topic_indices = [topic[0] for topic in topic_probabilities]
    return [topics_names[index] for index in topic_indices]

elicit_topic_name(doc_topics)
```
