```python
import pandas as pd 
fn= 'voted-kaggle-dataset.csv'
df = pd.read_csv(fn)
```

```python
print ('len of texts= {:,}'.format(len(df)))
df['Description']
print(type(df['Description'][4]))
len(df)
```

```python
df['Description']
```

```python
data=df['Description'].dropna()
len(data)
```

```python
from gensim import models
import gensim
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

three_words_pattern = r"\b\w{3,}\b"

stop  = list(set(stopwords.words('english')).union({'data', 'dataset', 'model','context','content'}))

vectorizer = CountVectorizer(
    min_df=20, 
    stop_words=stop,
    token_pattern=three_words_pattern) 
vectorizer.fit(data)
```

```python
print ('len of features = {:,}\n'.format(len(vectorizer.get_feature_names_out())))
print (vectorizer.get_feature_names_out()[:40])
```

```python
data_vectorized= vectorizer.transform(data)
print (data_vectorized)
```

```python
corpus = gensim.matutils.Sparse2Corpus(data_vectorized, documents_columns=False)
```

```python
id_map = dict((v, k) for k, v in vectorizer.vocabulary_.items())
id_map
```

```python
ldamodel = gensim.models.ldamodel.LdaModel (corpus, num_topics=10, id2word=id_map, passes=25, random_state=2024)
```

```python
ldamodel.print_topics(num_topics=10,num_words=30)
```
