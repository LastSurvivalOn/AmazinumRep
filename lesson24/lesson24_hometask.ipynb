```python
import pandas as pd 
fn= 'voted-kaggle-dataset.csv'
df = pd.read_csv(fn)
```

```python
print ('len of texts= {:,}'.format(len(df)))
df['Description']
print(type(df['Description'][4]))
len(df)
```

```python
data=df['Description'].dropna()
len(data)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle

three_words_pattern = r"\b\w{3,}\b"
vectorizer = CountVectorizer(
    min_df=20, 
    stop_words='english',
    token_pattern=three_words_pattern) 
vectorizer.fit(data)
```

```python
print ('len of features = {:,}\n'.format(len(vectorizer.get_feature_names_out())))
print (vectorizer.get_feature_names_out()[:40])
```

```python
newsgroup_data_vectorized= vectorizer.transform(data)
print (newsgroup_data_vectorized)
```

```python
corpus = gensim.matutils.Sparse2Corpus(newsgroup_data_vectorized, documents_columns=False)
# comparing to using corpora.Dictionary:
# corpus = [dictionary.doc2bow(text) for text in texts] 
[item for item in corpus][:5]
```

```python
id_map = dict((v, k) for k, v in vectorizer.vocabulary_.items()) 
id_map
```

```python
ldamodel = gensim.models.ldamodel.LdaModel (corpus, num_topics=5, id2word=id_map, passes=20, random_state=2024)
```

```python
ldamodel.print_topics(num_topics=5,num_words=60)
```

```python
topics_names= ['Sports', 'Social', 'Education', 'ML', 'Data']
```

```python
new_doc = ["\n\nThis team played well, match the best ",
           "\n\n\n\tThis big city is so danger and so criminal",
           "Ive just trained a model which recognizes on images the mood of person",
           "he downloaded ranking dataset from kaggle which contains csv files",
           "When i end the university i will go to college or take second speciality"]
print (new_doc[0])
```

```python
doc_vectorized = vectorizer.transform(new_doc)
new_doc_corpus = gensim.matutils.Sparse2Corpus(doc_vectorized, documents_columns=False)
doc_topics = ldamodel.get_document_topics(new_doc_corpus)
list(doc_topics)
```

```python
import numpy as np 
def elicit_topic_name(doc_topics):    
    return topics_names[np.squeeze(np.array(doc_topics))[:,1].argmax()]
for d in doc_topics:
    print(elicit_topic_name(d))
```
