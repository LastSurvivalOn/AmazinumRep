__Dataset description:__ Car insurance claim. As you are working in the insurance company. Company wants to know the reason why claim was not made. Doing so would allow insurance company to improve there policy for giving loan to the customer. In this project you are dealing with various feature such as age, occupation etc. based on that let's get back to the final conculsion.    

__Steps:__
- Data Cleaning
- Descriptive Statistics
- Data Visualization
- Conclusions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Car_Insurance_Claim.csv")
```

Data cleaning

```python
df = df.drop(df.columns[0], axis=1)
df
```

```python
df.replace('none', np.NaN, inplace=True)
df = df.dropna()
df = df.reset_index(drop=True)

df.shape
```

Descriptive statistics

```python
cols = df[['ID', 'CREDIT_SCORE', 'VEHICLE_OWNERSHIP', 'MARRIED', 'CHILDREN', 'POSTAL_CODE', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS', 'OUTCOME']]
cols = cols.astype(float)
cols.describe()
```

```python
str_cols = df[['AGE','GENDER','RACE','DRIVING_EXPERIENCE','EDUCATION','INCOME','VEHICLE_YEAR','VEHICLE_TYPE']]
str_cols = str_cols.describe(include='object')
str_cols
```

Data visualization

```python
cols.hist(figsize=(10, 10), bins=15, color="purple")
```

```python
colors = ["purple", "red", "yellow", "orange"]
fig, axes = plt.subplots(4, 2, figsize=(30, 30))
for i, col in enumerate(str_cols):
    values = df[col].value_counts()
    n_row = i // 2
    n_col = i % 2

    values.plot(kind="bar", ax=axes[n_row, n_col], color=colors[i % 4])
    axes[n_row, n_col].set_title(col)
    axes[n_row, n_col].set_xlabel(None)
```
